```python
!pip install bcrypt
import bcrypt

# Initialize an empty dictionary to store registered users
registered_users = {}

def hash_password(password):
    """
    Hashes a plain-text password using bcrypt.
    """
    hashed = bcrypt.hashpw(password.encode('utf-8'), bcrypt.gensalt())
    return hashed.decode('utf-8')

def check_password(plain_password, hashed_password):
    """
    Verifies a plain-text password against a hashed password using bcrypt.
    """
    try:
        return bcrypt.checkpw(plain_password.encode('utf-8'), hashed_password.encode('utf-8'))
    except ValueError:
        # Handle cases where hashed_password might be invalid or malformed
        return False

def register_user(username, password):
    """
    Registers a new user with input validation and password hashing.
    """
    # Input Validation
    if not username:
        print("Registration failed: Username cannot be empty.")
        return
    if not (3 <= len(username) <= 20):
        print("Registration failed: Username must be between 3 and 20 characters long.")
        return
    if not password:
        print("Registration failed: Password cannot be empty.")
        return
    if len(password) < 8:
        print("Registration failed: Password must be at least 8 characters long.")
        return

    # Uniqueness Check
    if username in registered_users:
        print(f"Registration failed: Username '{username}' is already taken.")
        return

    # Password Hashing and Storage
    hashed_pass = hash_password(password)
    registered_users[username] = hashed_pass
    print(f"User '{username}' registered successfully!")

def login_user(username, password):
    """
    Logs in a user with input validation and password verification.
    """
    # Input Validation
    if not username or not password:
        print("Login failed: Username and password cannot be empty.")
        return

    # User Verification
    if username not in registered_users:
        print("Invalid username or password.")
        return

    # Password Verification
    if check_password(password, registered_users[username]):
        print(f"Login successful for user: {username}!")
    else:
        print("Invalid username or password.")

def enhanced_secure_login_system():
    """
    Presents a menu for user registration and login.
    """
    print("--- Enhanced Secure Login System ---")
    while True:
        print("\nMenu:")
        print("1. Register")
        print("2. Login")
        print("3. Exit")

        choice = input("Enter your choice (1-3): ")

        if choice == '1':
            username = input("Enter new username: ")
            password = input("Enter new password: ")
            register_user(username, password)
        elif choice == '2':
            username = input("Enter username: ")
            password = input("Enter password: ")
            login_user(username, password)
        elif choice == '3':
            print("Exiting system. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter 1, 2, or 3.")

# Call the main function to run the system
enhanced_secure_login_system()
```

```python
#Task 2
def approve_loan(applicant_details):
    """
    Decides whether to approve or reject a loan based on applicant details.

    Args:
        applicant_details (dict):
A dictionary containing the applicant's details.
            Expected keys: 'credit_score', 'annual_income', 'loan_amount', 'debt_to_income_ratio'.

    Returns:
        str: 'Approved' if the loan meets criteria, 'Rejected' otherwise.
    """
    credit_score = applicant_details.get('credit_score')
    annual_income = applicant_details.get('annual_income')
    loan_amount = applicant_details.get('loan_amount')
    debt_to_income_ratio = applicant_details.get('debt_to_income_ratio')

    # Basic eligibility rules
    if credit_score is None or annual_income is None or loan_amount is None or debt_to_income_ratio is None:
        return "Rejected: Missing essential applicant details."

    if credit_score < 600:
        return "Rejected: Credit score too low."

    if annual_income < 30000:
        return "Rejected: Annual income too low."

    if loan_amount > (annual_income * 2):
        return "Rejected: Loan amount too high relative to income."

    if debt_to_income_ratio > 0.4:
        return "Rejected: Debt-to-income ratio too high."

    return "Approved"

# Example usage of the loan approval system
applicant1 = {
    'credit_score': 720,
    'annual_income': 60000,
    'loan_amount': 50000,
    'debt_to_income_ratio': 0.3
}

applicant2 = {
    'credit_score': 550,
    'annual_income': 45000,
    'loan_amount': 30000,
    'debt_to_income_ratio': 0.2
}

applicant3 = {
    'credit_score': 680,
    'annual_income': 40000,
    'loan_amount': 90000,
    'debt_to_income_ratio': 0.35
}

print(f"Applicant 1 loan status: {approve_loan(applicant1)}")
print(f"Applicant 2 loan status: {approve_loan(applicant2)}")
print(f"Applicant 3 loan status: {approve_loan(applicant3)}")
```

```python
#task 3
def recursive_binary_search(sorted_list, target, low=0, high=None):
    """
    Performs a binary search on a sorted list to find the target element recursively.

    Args:
        sorted_list (list): The list to search, must be sorted.
        target: The value to search for.
        low (int): The starting index of the current search segment (default is 0).
        high (int): The ending index of the current search segment (default is len(sorted_list) - 1).

    Returns:
        int: The index of the target if found, otherwise -1.
    """
    # Initialize high to the last index of the list if not provided (for the initial call)
    if high is None:
        high = len(sorted_list) - 1

    # Base Case 1: If the low index exceeds the high index, the target is not in the list.
    if low > high:
        return -1  # Target not found

    # Calculate the middle index of the current search segment.
    mid = (low + high) // 2

    # Base Case 2: If the element at the middle index is equal to the target, return the middle index.
    if sorted_list[mid] == target:
        return mid  # Target found at mid index

    # Recursive Case 1: If the element at the middle index is less than the target,
    # recursively call binary search on the right half of the list (from mid + 1 to high).
    elif sorted_list[mid] < target:
        return recursive_binary_search(sorted_list, target, mid + 1, high)

    # Recursive Case 2: If the element at the middle index is greater than the target,
    # recursively call binary search on the left half of the list (from low to mid - 1).
    else: # sorted_list[mid] > target
        return recursive_binary_search(sorted_list, target, low, mid - 1)

# Example usage of the recursive_binary_search function
sample_list = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
target1 = 23
target2 = 91
target3 = 10
target4 = 2

# Search for target1 (23)
index1 = recursive_binary_search(sample_list, target1)
if index1 != -1:
    print(f"Target {target1} found at index: {index1}")
else:
    print(f"Target {target1} not found in the list.")

# Search for target2 (91)
index2 = recursive_binary_search(sample_list, target2)
if index2 != -1:
    print(f"Target {target2} found at index: {index2}")
else:
    print(f"Target {target2} not found in the list.")

# Search for target3 (10 - not in list)
index3 = recursive_binary_search(sample_list, target3)
if index3 != -1:
    print(f"Target {target3} found at index: {index3}")
else:
    print(f"Target {target3} not found in the list.")

# Search for target4 (2)
index4 = recursive_binary_search(sample_list, target4)
if index4 != -1:
    print(f"Target {target4} found at index: {index4}")
else:
    print(f"Target {target4} not found in the list.")
```

```python
#Task 4
def score_applicant(applicant_data):
    """
    Calculates a score for a job applicant based on their skills, experience, and education.

    Args:
        applicant_data (dict): A dictionary containing applicant's information with keys:
            'skills' (list of str): List of relevant skills.
            'experience_years' (int): Years of relevant work experience.
            'education_level' (str): Highest education level (e.g., 'High School', 'Associate', 'Bachelor', 'Master', 'PhD').

            Note: This function explicitly avoids using any potentially biased features
            like 'name' or 'gender' in the scoring logic.

    Returns:
        int: The total score for the applicant.
    """

    score = 0

    # 1. Score based on Skills
    # Assign points for each relevant skill mentioned.
    # More granular scoring could involve weighting skills differently (e.g., critical vs. basic).
    skill_points_per_item = 5
    score += len(applicant_data.get('skills', [])) * skill_points_per_item

    # 2. Score based on Experience
    # Assign points based on years of experience. A linear or tiered system can be used.
    # Here, we'll use a simple linear model, capping at a certain point to avoid over-weighting.
    experience_years = applicant_data.get('experience_years', 0)
    experience_points_per_year = 10
    max_experience_score = 100 # Cap experience score
    score += min(experience_years * experience_points_per_year, max_experience_score)

    # 3. Score based on Education
    # Assign points based on the highest education level achieved.
    education_levels = {
        'High School': 10,
        'Associate': 20,
        'Bachelor': 40,
        'Master': 60,
        'PhD': 80
    }
    education_level = applicant_data.get('education_level', 'High School')
    score += education_levels.get(education_level, 0) # Default to 0 if education level is unknown

    return score

# --- Example Usage ---

# Define some sample applicants
applicant1 = {
    'skills': ['Python', 'SQL', 'Data Analysis', 'Machine Learning'],
    'experience_years': 5,
    'education_level': 'Master'
}

applicant2 = {
    'skills': ['Python', 'SQL'],
    'experience_years': 2,
    'education_level': 'Bachelor'
}

applicant3 = {
    'skills': ['JavaScript', 'HTML', 'CSS'],
    'experience_years': 8,
    'education_level': 'High School'
}

applicant4 = {
    'skills': ['Project Management', 'Communication'],
    'experience_years': 12,
    'education_level': 'PhD'
}

# Score the applicants
score1 = score_applicant(applicant1)
score2 = score_applicant(applicant2)
score3 = score_applicant(applicant3)
score4 = score_applicant(applicant4)

print(f"Applicant 1 Score: {score1}")
print(f"Applicant 2 Score: {score2}")
print(f"Applicant 3 Score: {score3}")
print(f"Applicant 4 Score: {score4}")
```

```python
#task 5
def process_employee_data(employee_details):
    """
    Processes employee details, intentionally including non-inclusive practices
    for analysis purposes.

    Args:
        employee_details (dict): A dictionary containing employee's details.
            Expected keys: 'employee_name', 'gender', 'marital_status', 'salary'.

    Returns:
        dict: A new dictionary with processed details including assumed
              'spouse_name' and 'yearly_bonus', along with original details.
    """
    processed_details = employee_details.copy()

    employee_name = employee_details.get('employee_name', 'N/A')
    gender = employee_details.get('gender', 'Unknown')
    marital_status = employee_details.get('marital_status', 'Unknown')
    salary = employee_details.get('salary', 0)

    # Non-inclusive practice 1: Assuming spouse based on gender and marital status
    processed_details['spouse_name'] = None
    if marital_status == 'Married' and gender == 'Male':
        # Intentionally non-inclusive: assuming a male employee has a wife
        processed_details['spouse_name'] = f"{employee_name.split()[-1]} (Wife's Name)"
    elif marital_status == 'Married' and gender == 'Female':
        # Intentionally non-inclusive: assuming a female employee has a husband
        processed_details['spouse_name'] = f"{employee_name.split()[-1]} (Husband's Name)"

    # Non-inclusive practice 2: Different bonus percentages based on gender
    yearly_bonus = 0
    if gender == 'Male':
        yearly_bonus = salary * 0.10 # 10% bonus for male
    elif gender == 'Female':
        yearly_bonus = salary * 0.08 # 8% bonus for female
    else:
        yearly_bonus = salary * 0.09 # Default 9% bonus for unknown/other gender

    processed_details['yearly_bonus'] = yearly_bonus

    return processed_details

# Example usage with sample employee details
employee1 = {
    'employee_name': 'John Doe',
    'gender': 'Male',
    'marital_status': 'Married',
    'salary': 75000
}

employee2 = {
    'employee_name': 'Jane Smith',
    'gender': 'Female',
    'marital_status': 'Married',
    'salary': 80000
}

employee3 = {
    'employee_name': 'Alex Johnson',
    'gender': 'Female',
    'marital_status': 'Single',
    'salary': 60000
}

employee4 = {
    'employee_name': 'Pat Lee',
    'gender': 'Other',
    'marital_status': 'Married',
    'salary': 90000
}

processed_employee1 = process_employee_data(employee1)
processed_employee2 = process_employee_data(employee2)
processed_employee3 = process_employee_data(employee3)
processed_employee4 = process_employee_data(employee4)

print("Processed Employee 1:", processed_employee1)
print("Processed Employee 2:", processed_employee2)
print("Processed Employee 3:", processed_employee3)
print("Processed Employee 4:", processed_employee4)
```

```python
def process_employee_data_inclusive(employee_details):
    """
    Processes employee details using inclusive and respectful coding practices.

    Args:
        employee_details (dict): A dictionary containing employee's details.
            Expected keys: 'employee_name', 'gender', 'marital_status', 'salary'.

    Returns:
        dict: A new dictionary with processed details including a generic
              'partner_details' and 'yearly_bonus', along with original details.
    """
    processed_details = employee_details.copy()

    employee_name = employee_details.get('employee_name', 'N/A')
    marital_status = employee_details.get('marital_status', 'Unknown')
    salary = employee_details.get('salary', 0)

    # Inclusive practice: Handling partner information without gender assumptions
    # We'll use a generic placeholder for partner's name if married.
    processed_details['partner_details'] = None
    if marital_status == 'Married':
        processed_details['partner_details'] = "[Partner's Name]"

    # Inclusive practice: Assigning a consistent bonus percentage to all employees
    # regardless of gender or any other protected characteristic.
    yearly_bonus_percentage = 0.10 # 10% bonus for all employees
    yearly_bonus = salary * yearly_bonus_percentage

    processed_details['yearly_bonus'] = yearly_bonus

    # Removing the old 'spouse_name' key if it exists from previous non-inclusive processing
    if 'spouse_name' in processed_details:
        del processed_details['spouse_name']

    return processed_details

# Example usage with sample employee details (re-using previous employee data)
# employee1, employee2, employee3, employee4 are already defined in the kernel state

processed_employee1_inclusive = process_employee_data_inclusive(employee1)
processed_employee2_inclusive = process_employee_data_inclusive(employee2)
processed_employee3_inclusive = process_employee_data_inclusive(employee3)
processed_employee4_inclusive = process_employee_data_inclusive(employee4)

print("Processed Employee 1 (Inclusive):", processed_employee1_inclusive)
print("Processed Employee 2 (Inclusive):", processed_employee2_inclusive)
print("Processed Employee 3 (Inclusive):", processed_employee3_inclusive)
print("Processed Employee 4 (Inclusive):", processed_employee4_inclusive)
```
