```python
class Student:
    """Represents a student with a name, roll number, and branch."""

    def __init__(self, name, roll_number, branch):
        """
        Initializes a new Student object.

        Args:
            name (str): The name of the student.
            roll_number (str): The roll number of the student.
            branch (str): The academic branch of the student.
        """
        self.name = name
        self.roll_number = roll_number
        self.branch = branch

    def display_details(self):
        """Prints the student's information in a readable format."""
        print(f"Student Name: {self.name}")
        print(f"Roll Number: {self.roll_number}")
        print(f"Branch: {self.branch}")

# --- After defining the class ---

# Create a Student object
print("Creating a student object...")
student1 = Student("Alice Smith", "CS101", "Computer Science")
print("Student object created successfully.\n")

# Call the display_details method to show the output on the console
print("Displaying student details:")
student1.display_details()

# You can create more student objects and display their details if needed
# print("\nCreating another student object...")
# student2 = Student("Bob Johnson", "EE205", "Electrical Engineering")
# print("Student object created successfully.\n")
# print("Displaying second student's details:")
# student2.display_details()
```

```python
def print_multiples_for(number):
    """Prints the first 10 multiples of a given number using a for loop."""
    print(f"First 10 multiples of {number} (using for loop):")
    # The range(1, 11) function generates numbers from 1 up to (but not including) 11.
    # So, it will iterate with values 1, 2, ..., 10.
    for i in range(1, 11):
        multiple = number * i
        print(multiple)

# Example output for number = 5
print_multiples_for(5)
```

```python
def print_multiples_while(number):
    """Prints the first 10 multiples of a given number using a while loop."""
    print(f"\nFirst 10 multiples of {number} (using while loop):")
    count = 1  # Initialize a counter variable
    # The loop continues as long as the 'count' is less than or equal to 10
    while count <= 10:
        multiple = number * count
        print(multiple)
        count += 1  # Increment the counter to eventually meet the termination condition

# Display the output for the while loop version
print_multiples_while(5)
```

```python
# --- Age Classification Function (using nested if-elif-else) ---

def classify_age_nested(age):
    """
    Classifies a person into age groups using nested if-elif-else statements.

    Age Group Definitions:
    - Child: 0-12 years
    - Teenager: 13-19 years
    - Adult: 20-64 years
    - Senior: 65+ years

    Args:
        age (int): The age of the person.

    Returns:
        str: The age group or an 'Invalid Age' message.
    """
    # First, check for invalid age (negative numbers). This forms the outer 'if' condition.
    if age < 0:
        return "Invalid Age: Age cannot be negative."
    else: # If age is not negative, proceed to classify into groups. This is the 'nested' part.
        # Check if the person is a Child
        if age <= 12:
            return "Child"
        # If not a Child, check if the person is a Teenager
        elif age <= 19:
            return "Teenager"
        # If not a Teenager, check if the person is an Adult
        elif age <= 64:
            return "Adult"
        # If none of the above, the person must be a Senior (65 or older)
        else:
            return "Senior"

# --- Example Outputs for Nested if-elif-else ---
print("--- Nested if-elif-else Classification ---")
print(f"Age 8: {classify_age_nested(8)}")       # Expected: Child
print(f"Age 16: {classify_age_nested(16)}")     # Expected: Teenager
print(f"Age 30: {classify_age_nested(30)}")     # Expected: Adult
print(f"Age 65: {classify_age_nested(65)}")     # Expected: Senior
print(f"Age -5: {classify_age_nested(-5)}\n")    # Expected: Invalid Age

# --- Explanation of Nested if-elif-else Logic ---
print("Explanation of Nested if-elif-else logic:")
print("------------------------------------------")
print("1. The function `classify_age_nested(age)` takes an integer `age` as input.")
print("2. **Outer `if` condition**: It first checks `if age < 0`. If true, it immediately returns 'Invalid Age' and the rest of the conditions are skipped.")
print("3. **`else` block (Nesting Point)**: If `age` is 0 or greater, the code enters the `else` block, which contains a series of `if-elif-else` statements. This is where the main classification happens.")
print("4. **First `if` in `else`**: It checks `if age <= 12`. If true, it returns 'Child'. If not, it moves to the next `elif`.")
print("5. **First `elif`**: It checks `elif age <= 19`. This condition is only evaluated if `age` was NOT `<= 12`. If true, it returns 'Teenager'. If not, it moves to the next `elif`.")
print("6. **Second `elif`**: It checks `elif age <= 64`. This condition is only evaluated if `age` was NOT `<= 19`. If true, it returns 'Adult'. If not, it moves to the final `else`.")
print("7. **Final `else`**: If `age` was greater than 64 (meaning it's not a Child, Teenager, or Adult), it defaults to returning 'Senior'.")
print("This flow ensures that each age falls into exactly one category based on the defined ranges, with invalid ages handled first.\n")

# --- Age Classification Function (using Dictionary-based/Mapping-based logic) ---

def classify_age_mapping(age):
    """
    Classifies a person into age groups using a dictionary-based (mapping) approach.

    Age Group Definitions:
    - Child: 0-12 years
    - Teenager: 13-19 years
    - Adult: 20-64 years
    - Senior: 65+ years

    Args:
        age (int): The age of the person.

    Returns:
        str: The age group or an 'Invalid Age' message.
    """
    # Handle invalid age first, similar to the nested approach for consistency.
    if age < 0:
        return "Invalid Age: Age cannot be negative."

    # Define age ranges and their corresponding groups using a list of (max_age, group_name) tuples.
    # This allows for easy iteration and checking.
    # The order matters here: check from youngest to oldest.
    age_ranges = [
        (12, "Child"),
        (19, "Teenager"),
        (64, "Adult"),
        (float('inf'), "Senior") # Use float('inf') for the upper bound of the last group
    ]

    # Iterate through the defined age ranges
    for max_age, group_name in age_ranges:
        # If the current age is less than or equal to the maximum age for this group,
        # then it belongs to this group.
        if age <= max_age:
            return group_name

    # This line should theoretically not be reached if float('inf') is used for Senior
    # but is good practice for exhaustive handling.
    return "Unknown"

# --- Example Outputs for Dictionary-based Classification ---
print("--- Dictionary-based Classification ---")
print(f"Age 8: {classify_age_mapping(8)}")       # Expected: Child
print(f"Age 16: {classify_age_mapping(16)}")     # Expected: Teenager
print(f"Age 30: {classify_age_mapping(30)}")     # Expected: Adult
print(f"Age 65: {classify_age_mapping(65)}")     # Expected: Senior
print(f"Age -5: {classify_age_mapping(-5)}\n")    # Expected: Invalid Age

# --- Comparison of the Two Approaches ---
print("Comparison of Age Classification Approaches:")
print("---------------------------------------------")
print("**1. Readability:**")
print("   - **Nested if-elif-else:** For simple, sequential conditions, it's quite readable for beginners as it directly maps to common logical thinking. However, if the 'nesting' becomes too deep or the conditions very complex, it can become harder to follow.")
print("   - **Dictionary-based (Mapping):** Can be very clean and readable, especially for defining the ranges. The lookup logic (the `for` loop) is concise once the `age_ranges` structure is understood. It separates the data (age ranges) from the logic.")
print("\n**2. Maintainability:**")
print("   - **Nested if-elif-else:** Modifying age ranges requires directly editing the conditional statements. If a new group is added, it might involve rearranging `elif` blocks carefully to maintain correct logic flow.")
print("   - **Dictionary-based (Mapping):** Highly maintainable. Age ranges can be easily updated or extended by modifying the `age_ranges` list without changing the core classification logic. This separation of concerns makes it very flexible.")
print("\n**3. Suitability for Beginners:**")
print("   - **Nested if-elif-else:** Excellent for beginners as `if`, `elif`, and `else` are fundamental concepts. The step-by-step logic is intuitive, though 'nesting' might introduce a slight conceptual hurdle if explained too literally as deep indentation.")
print("   - **Dictionary-based (Mapping):** A bit more advanced. While the concept of a list of tuples is straightforward, understanding iteration through it and checking conditions within a loop might require a slightly better grasp of Python data structures and loops. However, it's a great pattern to learn for more complex scenarios.")
print("\n**Conclusion:** For a small, fixed number of simple age groups, both are viable. The `if-elif-else` is perhaps more immediately accessible to absolute beginners. For systems requiring frequent changes to age categories or more complex grouping logic, the dictionary-based (mapping) approach offers superior maintainability and flexibility.")
```

### Implementation using a `for` loop

This approach iteratively adds each number from 1 to `n` to an accumulator variable.

```python
def sum_to_n_for_loop(n):
    """Calculates the sum of the first n natural numbers using a for loop."""
    if n < 1:
        return "Input must be a positive integer."

    total_sum = 0  # Initialize an accumulator variable
    print(f"Calculating sum of first {n} natural numbers using a for loop...")
    for i in range(1, n + 1):  # Loop from 1 up to and including n
        total_sum += i  # Add the current number to the total sum
        print(f"Adding {i}, Current sum: {total_sum}") # Show accumulation step-by-step
    return total_sum

# Example outputs
print(f"\nSum of first 5 natural numbers: {sum_to_n_for_loop(5)}\n")
print(f"Sum of first 10 natural numbers: {sum_to_n_for_loop(10)}\n")
print(f"Sum of first 0 natural numbers: {sum_to_n_for_loop(0)}")
```

```python
def sum_to_n_formula(n):
    """Calculates the sum of the first n natural numbers using a mathematical formula."""
    if n < 1:
        return "Input must be a positive integer."

    print(f"Calculating sum of first {n} natural numbers using a formula...")
    # The formula for the sum of the first n natural numbers is n * (n + 1) / 2
    result = n * (n + 1) // 2 # Use integer division for whole numbers
    return result

# Example outputs
print(f"\nSum of first 5 natural numbers: {sum_to_n_formula(5)}\n")
print(f"Sum of first 10 natural numbers: {sum_to_n_formula(10)}\n")
print(f"Sum of first 0 natural numbers: {sum_to_n_formula(0)}")
```

### `BankAccount` Class Implementation

This class simulates a basic bank account, allowing users to deposit, withdraw, and check their balance.

```python
class BankAccount:
    """Represents a simple bank account with deposit, withdrawal, and balance checking functionalities."""

    def __init__(self, initial_balance=0.0):
        """
        Initializes a new BankAccount object.

        Args:
            initial_balance (float): The starting balance of the account. Defaults to 0.0.
        """
        if initial_balance < 0:
            print("Warning: Initial balance cannot be negative. Setting to 0.0.")
            self.balance = 0.0
        else:
            self.balance = initial_balance
        print(f"Account created with initial balance: ${self.balance:.2f}")

    def deposit(self, amount):
        """
        Deposits money into the account.

        Args:
            amount (float): The amount of money to deposit.
        """
        if amount <= 0:
            print("Deposit amount must be positive.")
            return
        self.balance += amount
        print(f"Deposited: ${amount:.2f}. New balance: ${self.balance:.2f}")

    def withdraw(self, amount):
        """
        Withdraws money from the account.

        Args:
            amount (float): The amount of money to withdraw.
        """
        if amount <= 0:
            print("Withdrawal amount must be positive.")
            return
        if amount > self.balance:
            print(f"Insufficient funds. Attempted to withdraw ${amount:.2f}, but current balance is ${self.balance:.2f}.")
            return
        self.balance -= amount
        print(f"Withdrew: ${amount:.2f}. New balance: ${self.balance:.2f}")

    def check_balance(self):
        """
        Displays the current balance of the account.
        """
        print(f"Current account balance: ${self.balance:.2f}")

# --- Demonstration of the BankAccount Class ---
print("\n--- Demonstrating BankAccount ---")

# Create a new bank account object
my_account = BankAccount(100.00)

# Check initial balance
my_account.check_balance()

# Perform a deposit
my_account.deposit(50.75)

# Check balance after deposit
my_account.check_balance()

# Perform a withdrawal
my_account.withdraw(25.50)

# Check balance after withdrawal
my_account.check_balance()

# Attempt an invalid withdrawal (insufficient funds)
my_account.withdraw(200.00)

# Attempt an invalid deposit (negative amount)
my_account.deposit(-10.00)

# Check final balance
my_account.check_balance()

# Create another account with a zero initial balance
savings_account = BankAccount()
savings_account.deposit(1500.00)
savings_account.check_balance()
```
